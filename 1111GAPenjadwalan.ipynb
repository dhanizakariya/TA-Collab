```python
class Dosen :
  dosen = None

  def __init__(self,name,kode):
    self.name = name
    self.kode = kode

  @staticmethod
  def find(name):
    for i in range (len(Dosen.dosen)):
      if Dosen.dosen[i].name == name:
        return i
    return -1

  def __repr__(self):
    # return "Dosen: " + self.name
    return self.name

class Matkul :
  mamtkul = None

  def __init__(self,nomer,name, size, kode):
    self.name = name
    self.nomer = nomer
    # self.smt = smt
    self.size = size
    self.kode = kode

  @staticmethod
  def find(name):
    for i in range (len(Matkul.matkul)):
      if Matkul.matkul[i].name == name:
        return i
    return -1

  @staticmethod
  def find_by_kode(kode):
        for i in range(len(Matkul.matkul)):
            if Matkul.matkul[i].kode == kode:
                return i
        return -1

  def __repr__(self):
    # return "Matkul: " + self.name + ", Peserta: " + str(self.size)
    return self.name + ", \linebreak Peserta: " + str(self.size)


class Ruang :
  ruang = None

  def __init__(self,name, size, status = True):
    self.name = name
    self.size = size
    self.status = status

  @staticmethod
  def find(name):
    for i in range (len(Ruang.ruang)):
      if Ruang.ruang[i].name == name:
        return i
    return -1

  def __repr__(self):
    # return "Ruang: " + self.name + ", Kapasitas: " + str(self.size)
    return self.name + ", Kapasitas: " + str(self.size)

class Waktu :
  waktu = None

  def __init__(self, hari, mulai, selesai,  status_slot = False):
    self.mulai = mulai
    self.selesai = selesai
    self.hari = hari
    self.status_slot = status_slot

  def __repr__(self):
    # return "Hari: " + self.hari + ", Pukul: " + self.mulai + "-" + self.selesai
    return self.hari + " | " + self.mulai + "-" + self.selesai

```

```python
def create_individu():
  individu = [[random.randint(1, len(Dosen.dosen)),
              # i+1,
               random.randint(1, len(Matkul.matkul)),
               random.randint(1, len(Ruang.ruang)),
               random.randint(1, len(Waktu.waktu))
               ]
              for i in range (len(Matkul.matkul))]
  return individu
```

```python
def generate_matkul(x):
  a = random.choice(Dosen.dosen[x].kode)
  # print(a)
  matching_matkul = [matkul for matkul in Matkul.matkul if matkul.kode == a]
  # print(matching_matkul)
  selected_matkul = random.choice(matching_matkul)
  # print(selected_matkul)
  return selected_matkul.nomer
```

```python
def individu():
  individu = []
  for i in range(len(Matkul.matkul)):
    dosen = random.randint(1, len(Dosen.dosen))
    matkul = generate_matkul(dosen-1)
    ruang = random.randint(1, len(Ruang.ruang))
    waktu = random.randint(1, len(Waktu.waktu))
    a = [dosen,matkul,ruang,waktu]
    individu.append(a)
  return individu
```

```python
def _sort(sub_li,n):
    return(sorted(sub_li, key = lambda x: x[n]))
```

```python
def delete_duplicate(population, threshold=2):
    chromosome_counts = {}
    new_population = []

    for chromosome in population:
        chromosome_fitness = fitness(chromosome)
        if chromosome_fitness in chromosome_counts:
            chromosome_counts[chromosome_fitness] += 1
        else:
            chromosome_counts[chromosome_fitness] = 1

    for chromosome in population:
        chromosome_fitness = fitness(chromosome)
        if chromosome_counts[chromosome_fitness] <= threshold:
            new_population.append(chromosome)

    return new_population
```

```python
def elitism(population, new_chromosomes,len_pop):
    best = sorted(new_chromosomes, key=lambda x: fitness(x), reverse=True)
    if(fitness(best[0]) >= fitness(population[0])):
      population.append(best[0])

      sorted_population = sorted(population, key=lambda x: fitness(x), reverse=True)

      if (len(sorted_population) > len_pop):
        # remove_duplicate = delete_duplicate(sorted_population)
        # elite = remove_duplicate[:len_pop]
        elite = sorted_population[:len_pop]
      else :
        elite = sorted_population
    else:
      elite = population

    return elite,fitness(elite[0]),elite[0]
```

```python
def create_pop(n):
  pop = []
  for i in range (n):
    a = create_individu()
    pop.append(a)
  return pop
```

```python
#satu matkul hanya dilakukan satu kali
def cek_matkul(a):
  score = 0
  # a = _sort(a,1)
  for i in range (len(a)):
    for j in range (i+1, len(a)):
      if a[i][1] == a[j][1] :
        score = score + 1
  return score
```

```python
#satu ruangan hanya dipakai satu kali dalam satu waktu
def cek_ruang(a):
  score = 0
  # a = _sort(a,2)
  for i in range (len(a)):
    for j in range (i+1, len(a)):
      if (a[i][2] == a[j][2] and a[i][3] == a[j][3]):
        score = score + 1
  return score
```

```python
#satu dosen hanya bisa mengajar dalam satu waktu
def cek_dosen(a):
  score = 0
  for i in range (len(a)):
    for j in range (i+1, len(a)):
      if (a[i][0] == a[j][0] and a[i][3] == a[j][3] and Matkul.matkul[a[i][1]-1].kode != Matkul.matkul[a[j][1]-1].kode) :
        score = score+1
  return score
```

```python
def cek_kapasitas(a):
  score = 0
  for i in range (len(a)):
    for j in range (i+1, len(a)):
      if Matkul.matkul[a[i][1]-1].size > Ruang.ruang[a[i][2]-1].size:
        score = score + 1
  return score
```

```python
def fitness(a):
  scores = 0
  scores = scores + cek_matkul(a)
  # scores = scores + cek_dosen(a)
  # scores = scores + cek_ruang(a)
  # scores = scores + cek_kapasitas(a)
  return 1/((scores**2)  +1)
```

```python
def crossover(a,b):
  a = list(a)
  b = list(b)

  n = random.randint(1, (len(a)-1))
  m = random.randint(n, (len(a)-1))
  for i in range (n, m):
    a[i],b[i] = b[i],a[i]

  return a,b
```

```python
def mutate(a,b, mut_rate):
  n = len(a) + len(b)

  total_mutation = int(abs(mut_rate * n))
  print (total_mutation)
  if round(random.random(), 3) < mut_rate:
    print("Mutate\n")
    for i in range (total_mutation):
      dosen = random.randint(1, len(Dosen.dosen))
      matkul = generate_matkul(dosen-1)
      ruang = random.randint(1, len(Ruang.ruang))
      waktu = random.randint(1, len(Waktu.waktu))
      c = [dosen,matkul,ruang,waktu]
      x = random.randint(0,n)
      # print(x)
      if x <= (len(a)-1) :
        a[x] = c
      else :
        b[x-len(b)-1] = c
  else:
    print("Doesn't mutate")
  return a,b
```

```python
def selection(population):
    total_fitness = sum(fitness(chromosome) for chromosome in population)
    probabilities = [fitness(chromosome) / total_fitness for chromosome in population]

    selected_parents = []

    for _ in range(2):
        r = random.random()
        cumulative_probability = 0

        for i, probability in enumerate(probabilities):
            cumulative_probability += probability

            if cumulative_probability >= r:
                # print("individu nomer :", i)
                selected_parents.append(population[i])
                break

    return selected_parents[0],selected_parents[1]
```

```python
def gA(rate, limit, len_pop):
  pop = []
  score = 0.0
  generation = 0
  pop = create_pop(len_pop)
  while True:
    if generation == limit or score == 1.00:
      break
    else:
      print("Generation: ",generation)
      a,b= selection(pop)
      c,d = crossover(a,b)
      e,f = mutate(c,d,rate)
      pop,fit,best = elitism(pop,[c,d,e,f],len_pop)
      # print("jumlah populasi",len(pop))
      # best = pop[0]
      score = fit
      print("Fitness :", score)
      print("\n")
      generation = generation+1
  return best
```
